```python
#Importación de datos y librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "/content/drive/MyDrive/Bootcamp Coding Dojo/DataScience/Tareas Core/Archivos CSV-Tareas/retail_sales_dataset.csv"
df = pd.read_csv(url)
df
```

```python
# Filtrar datos inapropiados
df = df[(df['Age'] >= 18) & (df['Age'] <= 64)]
df = df[(df['Quantity'] >= 1) & (df['Quantity'] <= 4)]
df = df[(df['Price per Unit'] >= 25) & (df['Price per Unit'] <= 500)]
df = df[(df['Total Amount'] >= 25) & (df['Total Amount'] <= 2000)]
df
```

```python
# Estadísticas descriptivas
print(df.describe())
print(df['Gender'].value_counts())
print(df['Product Category'].value_counts())
```

```python
# Histograma y Boxplot de "Age"
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['Age'], bins=20, color='skyblue', kde=True)
plt.title('Distribución de Age')
plt.xlabel('Age')
plt.ylabel('Frecuencia')
plt.grid(True)

plt.subplot(1, 2, 2)
sns.boxplot(y=df['Age'], color='skyblue')
plt.title('Boxplot de Age')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

**Distribución de Age:**
* **Histograma:** La mayoría de los clientes tienen edades dentro del rango especificado.

* **Boxplot:** Confirma que los datos de edad están dentro de los límites, sin valores atípicos.


```python
# Histograma y Boxplot de "Quantity"
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['Quantity'], bins=4, color='lightgreen', kde=True)
plt.title('Distribución de Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frecuencia')
plt.grid(True)

plt.subplot(1, 2, 2)
sns.boxplot(y=df['Quantity'], color='lightgreen')
plt.title('Boxplot de Quantity')
plt.ylabel('Quantity')
plt.grid(True)
plt.show()
```

**Distribución de Quantity:**
* **Histograma:** La mayoría de las compras tienen una cantidad baja, lo cual es esperado.

* **Boxplot:** Muestra que los datos de cantidad están dentro de los límites, sin valores atípicos.

```python
# Histograma y Boxplot de "Total Amount"
plt.figure(figsize=(14, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['Total Amount'], bins=20, color='coral', kde=True)
plt.title('Distribución de Total Amount')
plt.xlabel('Total Amount')
plt.ylabel('Frecuencia')
plt.grid(True)

plt.subplot(1, 2, 2)
sns.boxplot(y=df['Total Amount'], color='coral')
plt.title('Boxplot de Total Amount')
plt.ylabel('Total Amount')
plt.grid(True)
plt.show()
```

**Distribución de Total Amount:**
* **Histograma:** La mayoría de las transacciones están dentro del rango especificado, con algunos valores más altos.

* **Boxplot:** Confirma la presencia de valores atípicos en transacciones de mayor cantidad, pero dentro del rango aceptable.

```python
# Gráficos de Líneas (Tendencias de Ventas)
# Convertir la columna 'Date' a tipo datetime
df['Date'] = pd.to_datetime(df['Date'])

# Agrupar por fecha y calcular el total de ventas por fecha
ventas_por_fecha = df.groupby('Date')['Total Amount'].sum().reset_index()

# Gráfico de líneas de ventas por fecha
plt.figure(figsize=(10, 6))
plt.plot(ventas_por_fecha['Date'], ventas_por_fecha['Total Amount'], marker='o')
plt.title('Tendencias de Ventas a lo Largo del Tiempo')
plt.xlabel('Fecha')
plt.ylabel('Total Amount')
plt.grid(True)
plt.show()
```

```python
# Gráficos de Dispersión (Relación entre Variables)
# Para detectar patrones correlativos

# Dispersión de Total Amount vs Quantity
plt.figure(figsize=(10, 6))
plt.scatter(df['Quantity'], df['Total Amount'], alpha=0.5)
plt.title('Total Amount vs Quantity')
plt.xlabel('Quantity')
plt.ylabel('Total Amount')
plt.grid(True)
plt.show()

# Dispersión de Age vs Total Amount
plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['Total Amount'], alpha=0.5)
plt.title('Age vs Total Amount')
plt.xlabel('Age')
plt.ylabel('Total Amount')
plt.grid(True)
plt.show()
```
