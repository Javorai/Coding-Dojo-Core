```python
#Importación de Datos y Librerias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
from google.colab import drive
drive.mount("/content/drive", force_remount=True)

# Ruta de la carpeta de los CSVs
archivos_drive = '/content/drive/MyDrive/Bootcamp Coding Dojo/DataScience/Tareas Core/Archivos CSV-Tareas/World Happiness Report'

# Encontrar todos los archivos CSV en la carpeta especificada
archivos_csv = glob.glob(archivos_drive + '/*.csv')

# Leer y unir todos los archivos CSV añadiendo la columna 'year' desde el nombre del archivo
dataframes = []
for archivo in archivos_csv:
    year = archivo.split('/')[-1].split('_')[1].split('.')[0]
    df_temp = pd.read_csv(archivo)
    df_temp['year'] = year
    dataframes.append(df_temp)

# Unir todos los DataFrames
df = pd.concat(dataframes, ignore_index=True)
df
```

```python
# Identificación de duplicados
duplicados = df.duplicated().sum()
print(f"Número de duplicados: {duplicados}")
```

```python
# Verificación de tipos de datos
df.info()
```

```python
#Detección de valores nulos
df.isnull().sum()
```

```python
#Limpieza de valores nulos
df = df.dropna()
print(df.isnull().sum())
```

```python
# Crear histogramas para las variables especificadas

# 1. happiness_score
plt.figure(figsize=(10, 6))
sns.histplot(df['happiness_score'], bins=20, color='skyblue', kde=True)
plt.title('Distribución de Happiness Score')
plt.xlabel('Happiness Score')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

# 2. gdp_per_capita
plt.figure(figsize=(10, 6))
sns.histplot(df['gdp_per_capita'], bins=20, color='lightgreen', kde=True)
plt.title('Distribución de GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

# 3. healthy_life_expectancy
plt.figure(figsize=(10, 6))
sns.histplot(df['healthy_life_expectancy'], bins=20, color='coral', kde=True)
plt.title('Distribución de Healthy Life Expectancy')
plt.xlabel('Healthy Life Expectancy')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()

# 4. freedom_to_make_life_choices
plt.figure(figsize=(10, 6))
sns.histplot(df['freedom_to_make_life_choices'], bins=20, color='violet', kde=True)
plt.title('Distribución de Freedom to Make Life Choices')
plt.xlabel('Freedom to Make Life Choices')
plt.ylabel('Frecuencia')
plt.grid(True)
plt.show()
```

```python
# Crear boxplots para las variables especificadas

# 1. happiness_score
plt.figure(figsize=(10, 6))
sns.boxplot(y=df['happiness_score'], color='skyblue')
plt.title('Boxplot de Happiness Score')
plt.ylabel('Happiness Score')
plt.grid(True)
plt.show()

# 2. gdp_per_capita
plt.figure(figsize=(10, 6))
sns.boxplot(y=df['gdp_per_capita'], color='lightgreen')
plt.title('Boxplot de GDP per Capita')
plt.ylabel('GDP per Capita')
plt.grid(True)
plt.show()

# 3. healthy_life_expectancy
plt.figure(figsize=(10, 6))
sns.boxplot(y=df['healthy_life_expectancy'], color='coral')
plt.title('Boxplot de Healthy Life Expectancy')
plt.ylabel('Healthy Life Expectancy')
plt.grid(True)
plt.show()

# 4. freedom_to_make_life_choices
plt.figure(figsize=(10, 6))
sns.boxplot(y=df['freedom_to_make_life_choices'], color='violet')
plt.title('Boxplot de Freedom to Make Life Choices')
plt.ylabel('Freedom to Make Life Choices')
plt.grid(True)
plt.show()
```

# **ANALISIS DE RESULTADOS**


```python
plt.figure(figsize=(14, 6))

# Histograma
plt.subplot(1, 2, 1)
sns.histplot(df['happiness_score'], bins=20, color='skyblue', kde=True)
plt.title('Distribución de Happiness Score')
plt.xlabel('Happiness Score')
plt.ylabel('Frecuencia')

# Boxplot
plt.subplot(1, 2, 2)
sns.boxplot(y=df['happiness_score'], color='skyblue')
plt.title('Boxplot de Happiness Score')
plt.ylabel('Happiness Score')

plt.show()
```

**Distribución del Happiness Score:**

* **Histograma:** La mayoría de los países tienen un puntaje de felicidad entre 4 y 7, con una distribución bastante normal.

* **Boxplot:** Revela algunos valores atípicos con puntajes muy altos o muy bajos, indicando países con niveles de felicidad extremadamente altos o bajos.

```python
plt.figure(figsize=(14, 6))

# Histograma
plt.subplot(1, 2, 1)
sns.histplot(df['gdp_per_capita'], bins=20, color='lightgreen', kde=True)
plt.title('Distribución de GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Frecuencia')

# Boxplot
plt.subplot(1, 2, 2)
sns.boxplot(y=df['gdp_per_capita'], color='lightgreen')
plt.title('Boxplot de GDP per Capita')
plt.ylabel('GDP per Capita')

plt.show()
```
